### Features with low variance

Some features have only one value. Therefore, they add no information, and should be dropped.

```python
features_with_one_value = [] 
for column in df.columns:
    if len(df[column].value_counts())==1:
        print(f"{column} has only one value")
        features_with_one_value.append(column)

df.drop(columns=features_with_one_value, inplace=True)
```

### correlation

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corr = df.corr()
```

```python
plt.figure(figsize = (12,12))
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    linewidths=0
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
